```
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import os
import re

# Mostra o diretório atual e os arquivos na pasta datasets
print("Diretório atual:", os.getcwd())
print("Arquivos na pasta datasets:", os.listdir("../datasets"))
# todo executar novamente a tranformacao para teste
# todo testar outras formas de limpeza...
# Leitura do CSV
try:
    df = pd.read_csv("../datasets/phishing.csv", encoding="utf-8")
except UnicodeDecodeError:
    df = pd.read_csv("../datasets/phishing.csv", encoding="ISO-8859-1")

df
```

```
# Verifica colunas não numéricas
colunas_nao_numericas = df.select_dtypes(include=['object']).columns.tolist()
colunas_nao_numericas
```

```
df = pd.get_dummies(df, columns=['Email Type'])
df
```

```
df_transfomado = df.drop(columns=['Email Type_Safe Email'])
df_transfomado
```

```
resultado = df_transfomado['Email Text'].apply(
    lambda x: re.sub(r'[^a-zA-Z0-9áéíóúãõâêôçÁÉÍÓÚÃÕÂÊÔÇ\s]', '', str(x))
)
df_transfomado['Email Text'] = resultado
# Criar colunas binárias para cada palavra
df_dummies = df_transfomado['Email Text'].str.lower().str.get_dummies(sep=' ')

# Juntar no dataframe original
df_transfomado = df_transfomado.join(df_dummies)
df_transfomado = df_transfomado.drop(columns=['Email Text'])
df_transfomado.to_csv("datasets/phishing_transformed.csv", index=False)
```
